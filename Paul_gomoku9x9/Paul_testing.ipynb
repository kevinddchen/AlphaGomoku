## Older versus younger

```python
model13 = player_zero.net(SIZE, l2=1e-5)
model11 = player_zero.net(SIZE, l2=1e-5)
model13.load_weights('day13.h5')
model11.load_weights('day11.h5')
```

```python
wincount = 0
for i in range(20):
    game = gomoku.Gomoku(SIZE)
    p1 = player_zero.ZeroPlayer('p1', +1, game, model11)
    p2 = player_zero.ZeroPlayer('p2', -1, game, model13)
    pa, pb = p1, p2
    while not game.finished:
        game.play(*pa.play(game, N_ITER))
        pa, pb = pb, pa
    game.show()
    if game.winner == -1:
        wincount += 1
print ('Won '+str(wincount)+' games out of 20 games as white')
```

```python
wincount = 0
for i in range(20):
    game = gomoku.Gomoku(SIZE)
    p1 = player_zero.ZeroPlayer('p1', +1, game, model13)
    p2 = player_zero.ZeroPlayer('p2', -1, game, model11)
    pa, pb = p1, p2
    while not game.finished:
        game.play(*pa.play(game, N_ITER))
        pa, pb = pb, pa
    game.show()
    if game.winner == 1:
        wincount += 1
print ('Won '+str(wincount)+' games out of 20 games as black')
```

# Play agains Feature player

```python
model13 = player_zero.net(SIZE, l2=1e-5)
model13.load_weights('day13.h5')
```

```python
wincount = 0
for i in range(20):
    game = gomoku.Gomoku(SIZE)
    p1 = player_zero.ZeroPlayer('p1', +1, game, model13)
    p2 = player_basic.FeaturePlayer('p2', -1)
    while not game.finished:
        game.play(*p1.play(game, N_ITER))
        if not game.finished:
            game.play(*p2.play(game))
    game.show()
    if game.winner == 1:
        wincount += 1
print ('Won '+str(wincount)+' games out of 20 games as black agains FeaturePlayer')
```

```python
wincount = 0
for i in range(20):
    game = gomoku.Gomoku(SIZE)
    p2 = player_zero.ZeroPlayer('p2', -1, game, model13)
    p1 = player_basic.FeaturePlayer('p1', +1)
    while not game.finished:
        game.play(*p1.play(game))
        if not game.finished:
            game.play(*p2.play(game,N_ITER))
    game.show()
    if game.winner == -1:
        wincount += 1
print ('Won '+str(wincount)+' games out of 20 games as white agains FeaturePlayer')
```
